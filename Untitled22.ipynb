```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data[:, 0].reshape(-1, 1)  # Sepal length (input), reshaped to fit scikit-learn
y = iris.data[:, 1]  # Sepal width (what we predict)

# Model
model = LinearRegression()  # Creates a blank slate for a straight line
model.fit(X, y)  # Finds the best line (y = mx + b) by adjusting m (slope) and b (intercept)

# Predict
y_pred = model.predict(X)  # Uses the line to guess sepal width for each sepal length

# Graph
plt.scatter(X, y, color='blue', label='Data')  # Plots real data points
plt.plot(X, y_pred, color='red', label='Linear Fit')  # Draws the line
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import numpy as np

# Load data
iris = load_iris()
X = iris.data[:, [0, 1]]  # Sepal length and width (two inputs)
y = iris.target  # Classes (0, 1, 2)

# Model
model = DecisionTreeClassifier(max_depth=3)  # Sets up a tree with max 3 levels
model.fit(X, y)  # Builds the tree by splitting data

# Graph
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')  # Colors points by class
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Decision Tree Classification')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import numpy as np

# Load data
iris = load_iris()
X = iris.data[:, [0, 1]]  # Sepal length and width
y = iris.target

# Model
model = LogisticRegression(max_iter=200)  # Probability-based classifier
model.fit(X, y)  # Finds separating lines

# Graph
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Logistic Regression')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_iris
import numpy as np

# Load data
iris = load_iris()
X = iris.data[:, [0, 1]]  # Sepal length and width
y = iris.target

# Model
model = SVC(kernel='linear')  # Linear separator
model.fit(X, y)  # Finds the widest-gap line

# Graph
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Support Vector Machine')
plt.show()
```
