```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data[:, 0].reshape(-1, 1)  # Sepal length
y = iris.data[:, 1]  # Sepal width

# Model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)

# Graph
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Linear Fit')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data[:, [0, 1]]  # Sepal length and width
y = iris.target

# Model
model = DecisionTreeClassifier(max_depth=3)
model.fit(X, y)

# Create mesh grid for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Graph
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')  # Decision regions
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', label='Data')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Decision Tree Classification')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data[:, [0, 1]]  # Sepal length and width
y = iris.target

# Model
model = LogisticRegression(max_iter=200)
model.fit(X, y)

# Mesh grid for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Graph
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')  # Decision regions
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', label='Data')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Logistic Regression')
plt.show()
```
